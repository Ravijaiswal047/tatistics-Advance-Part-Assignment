```python
# --------------------PRACTICAL---------------------------

1.  Write a Python program to generate a random variable and display its value?
sol.  import random
     random_variable = random.randint(1,100)
     print(random_variable)

2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?
sol. import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of the discrete uniform distribution
a, b = 1, 10

# Generate random samples
samples = randint.rvs(a, b+1, size=1000)

# Get the PMF values
x_values = np.arange(a, b+1)
pmf_values = [1 / (b - a + 1)] * len(x_values)

# Plot the PMF
plt.bar(x_values, pmf_values, alpha=0.7, color="skyblue", edgecolor="black", label="PMF")

# Labels and title
plt.xlabel("Random Variable (X)")
plt.ylabel("Probability")
plt.title("Discrete Uniform Distribution PMF")
plt.xticks(x_values)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
sol. import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

def bernoulli_pmf(p):

    # Possible values of X in Bernoulli (0 or 1)
    x_values = [0, 1]

    # Compute PMF values
    pmf_values = [1 - p, p]  # P(X=0) = 1-p, P(X=1) = p

    # Plot the PMF
    plt.bar(x_values, pmf_values, color='skyblue', edgecolor='black', width=0.4)

    # Labels and title
    plt.xticks([0, 1])
    plt.xlabel("Random Variable (X)")
    plt.ylabel("Probability")
    plt.title(f"Bernoulli Distribution PMF (p={p})")

4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10   # Number of trials
p = 0.5  # Probability of success

# Simulate 1000 samples from the binomial distribution
samples = np.random.binomial(n, p, size=1000)

# Plot histogram
plt.hist(samples, bins=np.arange(12) - 0.5, density=True, alpha=0.6, color='skyblue', edgecolor='black', label="Simulated Data")

# Plot the theoretical PMF
x = np.arange(0, n + 1)
pmf_values = binom.pmf(x, n, p)
plt.plot(x, pmf_values, 'ro-', label="Theoretical PMF")

# Labels and title
plt.xlabel("Number of Successes (X)")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution Histogram (n={n}, p={p})")
plt.xticks(range(n + 1))
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()

5.Create a Poisson distribution and visualize it using Python?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the rate parameter (lambda)
lam = 5  # Average number of events in a given interval

# Generate Poisson-distributed data
samples = poisson.rvs(mu=lam, size=1000)

# Plot histogram of simulated data
plt.hist(samples, bins=np.arange(15) - 0.5, density=True, alpha=0.6, color='skyblue', edgecolor='black', label="Simulated Data")

# Compute and plot the theoretical PMF
x = np.arange(0, 15)
pmf_values = poisson.pmf(x, mu=lam)
plt.plot(x, pmf_values, 'ro-', label="Theoretical PMF")

# Labels and title
plt.xlabel("Number of Events (X)")
plt.ylabel("Probability")
plt.title(f"Poisson Distribution (λ={lam})")
plt.xticks(range(15))
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of the discrete uniform distribution
a, b = 1, 10  # Lower and upper bounds

# Generate possible values
x_values = np.arange(a, b + 1)

# Compute CDF values
cdf_values = randint.cdf(x_values, a, b + 1)

# Plot the CDF
plt.step(x_values, cdf_values, where="mid", color="blue", label="CDF", linewidth=2)

# Labels and title
plt.xlabel("Random Variable (X)")
plt.ylabel("Cumulative Probability")
plt.title(f"CDF of Discrete Uniform Distribution (a={a}, b={b})")
plt.xticks(x_values)
plt.yticks(np.linspace(0, 1, b - a + 2))
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.legend()

# Show the plot
plt.show()

7.Generate a continuous uniform distribution using NumPy and visualize it?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Define the range of the uniform distribution
a, b = 2, 10  # Lower and upper bounds

# Generate random samples
samples = np.random.uniform(a, b, size=1000)

# Generate values for PDF
x_values = np.linspace(a - 1, b + 1, 1000)  # Extend range slightly for visibility
pdf_values = uniform.pdf(x_values, a, b - a)

# Plot histogram of samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label="Simulated Data")

# Plot theoretical PDF
plt.plot(x_values, pdf_values, 'r-', label="

8.Simulate data from a normal distribution and plot its histogram?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters for the normal distribution
mu, sigma = 50, 10  # Mean and standard deviation

# Generate random samples from a normal distribution
samples = np.random.normal(mu, sigma, 1000)

# Plot histogram of samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label="Simulate

9. Write a Python function to calculate Z-scores from a dataset and plot them
  import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_z_scores(data):
    """
    Calculates Z-scores from a dataset and plots them.

    Parameters:
        data (list or numpy array): The dataset

    Returns:
        None (displays a plot)
    """
    # Compute mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Compute Z-scores
    z_scores = (data - mean) / std_dev

    # Plot Z-scores
    plt.figure(figsize=(10, 5))
    sns.histplot(z_scores, bins=30, kde=True, color="skyblue", edgecolor="black")

    # Add reference lines at mean (0) and ±1, ±2, ±3 standard deviations
    for z in [-3, -2, -1, 0, 1, 2, 3]:
        plt.axvline(z, color='red' if z != 0 else 'black', linestyle="--", alpha=0.7, label=f"Z={z}" if z != 0 else "Mean (Z=0)")

    # Labels and title
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.title("Z-score Distribution")
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Show the plot
    plt.show()

# Example usage
data = np.random.normal(50, 10, 1000)  # Generate normal distribution (mean=50, std=10)
plot_z_scores(data)

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def demonstrate_clt(sample_size=30, num_samples=1000):

    # Step 1: Define a Non-Normal Distribution (Uniform)
    population = np.random.uniform(0, 100, 100000)  # Large uniform distribution

    # Step 2: Draw multiple random samples and compute their means
    sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

    # Step 3: Plot the histogram of sample means
    plt.figure(figsize=(10, 5))
    sns.histplot(sample_means, bins=30, kde=True, color="skyblue", edgecolor="black")

    # Labels and title
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.title(f"Central Limit Theorem Demonstration\nSample Size={sample_size}, Num Samples={num_samples}")

    # Show the plot
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

# Example Usage
demonstrate_clt(sample_size=30, num_samples=1000)

15.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def verify_clt_normal(pop_mean=50, pop_std=10, sample_size=30, num_samples=1000):

    # Step 1: Generate a large normal population
    population = np.random.normal(pop_mean, pop_std, 100000)

    # Step 2: Draw multiple random samples and compute their means
    sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

    # Step 3: Plot histogram of sample means
    plt.figure(figsize=(10, 5))
    sns.histplot(sample_means, bins=30, kde=True, color="skyblue", edgecolor="black")

    # Plot theoretical normal distribution for sample means
    theoretical_std = pop_std / np.sqrt(sample_size)
    x_values = np.linspace(pop_mean - 4 * theoretical_std, pop_mean + 4 * theoretical_std, 1000)
    normal_curve = (1 / (theoretical_std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_values - pop_mean) / theoretical_std) ** 2)
    plt.plot(x_values, normal_curve * num_samples * (sample_size / 10), 'r-', label="Theoretical Normal Curve", linewidth=2)

    # Labels and title
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.title(f"Verification of CLT (Sample Size={sample_size}, Num Samples={num_samples})")
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Show the plot
    plt.show()

# Example Usage
verify_clt_normal(sample_size=30, num_samples=1000)

16.Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal():
    """
    Calculates and plots the Standard Normal Distribution (mean = 0, std = 1).
    """
    # Generate values for X-axis
    x_values = np.linspace(-4, 4, 1000)

    # Compute PDF of standard normal distribution
    pdf_values = norm.pdf(x_values, 0, 1)

    # Plot the standard normal curve
    plt.plot(x_values, pdf_values, 'b-', label="Standard Normal PDF", linewidth=2)

    # Add vertical lines for mean and standard deviations
    for z in [-3, -2, -1, 0, 1, 2, 3]:
        plt.axvline(z, color='red' if z != 0 else 'black', linestyle="--", alpha=0.7, label=f"Z={z}" if z != 0 else "Mean (Z=0)")

    # Labels and title
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Standard Normal Distribution (μ=0, σ=1)")
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Show the plot
    plt.show()

# Example Usage
plot_standard_normal()

17.Generate random variables and calculate their corresponding probabilities using the binomial distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def binomial_distribution(n=10, p=0.5, size=1000):

    # Generate binomial random variables
    samples = np.random.binomial(n, p, size)

    # Calculate theoretical PMF values
    x_values = np.arange(0, n + 1)
    pmf_values = binom.pmf(x_values, n, p)

    # Plot histogram of simulated data
    plt.hist(samples, bins=np.arange(n+2)-0.5, density=True, alpha=0.6, color='skyblue', edgecolor='black', label="Simulated Data")

    # Plot theoretical PMF
    plt.plot(x_values, pmf_values, 'ro-', label="Theoretical PMF", linewidth=2)

    # Labels and title
    plt.xlabel("Number of Successes (X)")
    plt.ylabel("Probability")
    plt.title(f"Binomial Distribution (n={n}, p={p})")
    plt.xticks(range(n + 1))
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Show the plot
    plt.show()

# Example Usage
binomial_distribution(n=10, p=0.5, size=1000)

18.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution
         import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):

    # Compute mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-score
    z_score = (x - mean) / std_dev

    # Generate standard normal distribution curve
    x_values = np.linspace(-4, 4, 1000)
    pdf_values = norm.pdf(x_values, 0, 1)

    # Plot the standard normal distribution
    plt.plot(x_values, pdf_values, 'b-', label="Standard Normal Distribution (μ=0, σ=1)", linewidth=2)

    # Highlight the Z-score on the plot
    plt.axvline(z_score, color='red', linestyle="--", label=f"Z-score for x={x}: {z_score:.2f}")

    # Labels and title
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Z-score Comparison with Standard Normal Distribution")
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Show the plot
    plt.show()

    return z_score

# Example Usage
data = np.random.normal(50, 10, 1000)  # Generate normal dataset (mean=50, std=10)
x_value = 60  # Data point to compare
z = calculate_z_score(data, x_value)

print(f"The Z-score for x={x_value} is {z:.2f}")

19. 5 Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):

    # Step 1: Compute sample mean and size
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Step 2: Compute the Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Step 3: Find the critical Z value (two-tailed test)
    z_critical = norm.ppf(1 - alpha / 2)

    # Step 4: Decision Rule
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.2f}")
    print(f"Critical Z-value: ±{z_critical:.2f}")

    if abs(z_score) > z_critical:
        print(" Reject the Null Hypothesis (H0): Significant difference found.")
    else:
        print(" Fail to Reject the Null Hypothesis (H0): No significant difference found.")

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(50, 10, 100)  # Generate normal dataset (mean=50, std=10, n=100)
z_test(sample_data, pop_mean=52, pop_std=10, alpha=0.05)


20.Create a confidence interval for a dataset using Python and interpret the result
         import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence_level=0.95):

    # Step 1: Calculate sample mean, sample standard deviation, and sample size
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data)

    # Step 2: Calculate the critical Z-value for the given confidence level
    alpha = 1 - confidence_level
    z_critical = norm.ppf(1 - alpha / 2)  # Z-value for two-tailed test

    # Step 3: Calculate the margin of error
    margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))

    # Step 4: Calculate confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(50, 10, 100)  # Sample data (mean=50, std=10, n=100)

# Calculate confidence interval at 95% confidence level
lower, upper = confidence_interval(sample_data, confidence_level=0.95)

print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")

21.Create a confidence interval for a dataset using Python and interpret the result
         import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence_level=0.95):
    """
    Calculates the confidence interval for the given data.

    Parameters:
        data (list or numpy array): The dataset
        confidence_level (float): Desired confidence level (default is 0.95)

    Returns:
        tuple: (lower bound, upper bound) of the confidence interval
    """
    # Step 1: Calculate sample mean, sample standard deviation, and sample size
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data)

    # Step 2: Calculate the critical Z-value for the given confidence level
    alpha = 1 - confidence_level
    z_critical = norm.ppf(1 - alpha / 2)  # Z-value for two-tailed test

    # Step 3: Calculate the margin of error
    margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))

    # Step 4: Calculate confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(50, 10, 100)  # Sample data (mean=50, std=10, n=100)

# Calculate confidence interval at 95% confidence level
lower, upper = confidence_interval(sample_data, confidence_level=0.95)

print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")

21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence_level=0.95):

    # Step 1: Calculate sample mean, sample standard deviation, and sample size
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data)

    # Step 2: Calculate the critical Z-value for the given confidence level
    alpha = 1 - confidence_level
    z_critical = norm.ppf(1 - alpha / 2)  # Z-value for two-tailed test

    # Step 3: Calculate the margin of error
    margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))

    # Step 4: Calculate confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Generate data from a normal distribution (mean=50, std=10, n=100)
np.random.seed(42)
data = np.random.normal(50, 10, 100)  # Sample size = 100

# Calculate the 95% confidence interval for the mean
lower, upper = confidence_interval(data, confidence_level=0.95)

print(f"95% Confidence Interval for the mean: ({lower:.2f}, {upper:.2f})")

22. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mu=0, sigma=1):

    # Generate a range of x values
    x_values = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

    # Calculate the PDF values for the given range of x
    pdf_values = norm.pdf(x_values, mu, sigma)

    # Plot the PDF
    plt.plot(x_values, pdf_values, label=f"Normal Distribution (μ={mu}, σ={sigma})", color='blue')

    # Labels and title
    plt.xlabel("x")
    plt.ylabel("Probability Density")
    plt.title("Probability Density Function of a Normal Distribution")
    plt.legend()

    # Display grid
    plt.grid(True)

    # Show the plot
    plt.show()

# Example usage: Plot the PDF for a standard normal distribution (mean=0, std=1)
plot_normal_pdf(mu=0

23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
  import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lmbda=5, k_max=15):

    # Generate x values (0 to k_max)
    x_values = np.arange(0, k_max + 1)

    # Calculate the CDF values for the Poisson distribution
    cdf_values = poisson.cdf(x_values, lmbda)

    # Plot the CDF
    plt.step(x_values, cdf_values, where='mid', label=f'Poisson CDF (λ={lmbda})', color='blue', linewidth=2)

    # Labels and title
    plt.xlabel("Number of Events (k)")
    plt.ylabel("Cumulative Probability")
    plt.title("Cumulative Distribution Function of Poisson Distribution")
    plt.legend()

    # Display grid
    plt.grid(True)

    # Show the plot
    plt.show()

24. Simulate a random variable using a continuous uniform distribution and calculate its expected value
 import numpy as np
import matplotlib.pyplot as plt

def simulate_uniform_distribution(a, b, size=1000):

    # Step 1: Simulate random samples from the continuous uniform distribution
    samples = np.random.uniform(a, b, size)

    # Step 2: Calculate the expected value (mean) of the uniform distribution
    expected_value = (a + b) / 2

    # Step 3: Plot the histogram of the simulated data
    plt.hist(samples, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')

    # Add a vertical line for the expected value
    plt.axvline(expected_value, color='red', linestyle='dashed', linewidth=2, label=f"Expected Value: {expected_value:.2f}")

    # Labels and title
    plt.xlabel("Random Variable (X)")
    plt.ylabel("Probability Density")
    plt.title(f"Continuous Uniform Distribution (a={a}, b={b})")
    plt.legend()

    # Show the plot
    plt.show()

    return expected_value

# Example usage: Simulate a uniform distribution with a=10, b=20
a = 10
b = 20
expected_value = simulate_uniform_distribution(a, b)
print(f"Expected Value of the distribution: {expected_value}")

25 Write a Python program to compare the standard deviations of two datasets and visualize the difference?
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):

    # Step 1: Calculate the standard deviations of both datasets
    std_dev1 = np.std(data1, ddof=1)  # Sample standard deviation (ddof=1 for sample)
    std_dev2 = np.std(data2, ddof=1)

    # Step 2: Compare standard deviations
    print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}
26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np

def calculate_range_iqr(data):

    # Calculate the range
    data_range = np.max(data) - np.min(data)

    # Calculate the IQR (Interquartile Range)
    Q1 = np.percentile(data, 25)  # 25th percentile
    Q3 = np.percentile(data, 75)  # 75th percentile
    IQR = Q3 - Q1

    return data_range, IQR

# Generate a dataset from a normal distribution (mean=50, std=10, n=1000)
np.random.seed(42)
data = np.random.normal(50, 10, 1000)

# Calculate range and IQR for the dataset
data_range, IQR = calculate_range_iqr(data)

print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {IQR:.2f}")


27. Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Apply Z-score normalization
    normalized_data = (data - mean) / std_dev

    return normalized_data

# Example usage: Generate a dataset from a normal distribution (mean=50, std=10, n=1000)
np.random.seed(42)
data = np.random.normal(50, 10, 1000)

# Normalize the dataset using Z-score normalization
normalized_data = z_score_normalization(data)

# Plot the original and normalized data
plt.figure(figsize=(12, 6))

# Plot the original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the normalized data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='orange', edgecolor='black')
plt.title('Z-score Normalized Data Distribution')
plt.xlabel('Normalized Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):

    # Calculate skewness
    data_skewness = skew(data)

    # Calculate kurtosis (Fisher's definition: subtract 3 to make normal distribution have kurtosis = 0)
    data_kurtosis = kurtosis(data)

    return data_skewness, data_kurtosis

# Example usage: Generate a dataset from a normal distribution (mean=50, std=10, n=1000)
np.random.seed(42)
data = np.random.normal(50, 10, 1000)

# Calculate skewness and kurtosis
skewness, kurt = calculate_skewness_kurtosis(data)

print(f"Skewness of the dataset: {skewness:.4f}")
print(f"Kurtosis of the dataset: {kurt:.4f}")
```
